```python
import re
import html

def text_to_html_inline(text: str) -> str:
    """
    Convert plain text into <p>paragraph</p> blocks, inline.
    Paragraphs are separated by blank lines.
    """
    # Normalize line endings
    text = text.replace('\r\n', '\n').replace('\r', '\n')

    # Split into paragraphs on blank lines
    paras = [p.strip() for p in re.split(r'\n\s*\n+', text) if p.strip()]

    # Escape HTML special chars and collapse whitespace
    paras = [re.sub(r'\s+', ' ', html.escape(p)).strip() for p in paras]

    # Wrap each in <p> ... </p>, inline
    return "\n".join(f'<p style="text-indent: 2em; margin-bottom: 20em;">{p}</p>' for p in paras)

# --- Apply to your file barren.txt ---
with open("barren.txt", "r", encoding="utf-8") as f:
    text = f.read()

html_output = text_to_html_inline(text)
print(html_output)
```
